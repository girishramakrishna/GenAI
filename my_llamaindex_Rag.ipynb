```python
from langchain import PromptTemplate, LLMChain

question="Who won the Cricket World Cup in the year 2011?"
template = """Question: {question}
Answer: Let's think step by step."""
prompt = PromptTemplate(template=template, input_variables=["question"])
print(prompt)
```

```python
# LLM chain
llm_chain=LLMChain(llm=llm,prompt=prompt)
# Input params: question
print(llm_chain.invoke(question))
```

```python

from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext
#from llama_index.llms import HuggingFaceLLM
```

```python
# Loading Documents
documents=SimpleDirectoryReader("/content").load_data(show_progress=True)
```

```python
# No. of documents loaded
doc_count=len(documents)
print(f"Document count: {doc_count}")
```

```python
# Display first 3 documents
documents[:3]
```

```python
index=VectorStoreIndex.from_documents(documents, show_progress= True)
```

```python
#!pip install llama-index-llms-langchain
```

```python
query_engine=index.as_query_engine(llm=llm)
```

```python
response1=query_engine.query("What is Nestlé’s approach to ensuring a safe and healthy work environment?")
```

```python
print(response1)
```

```python
response2=query_engine.query("What is Expatriation Policy of Nestle")
print(response2)
```

```python
response2=query_engine.query("What is Corporate Policy of Nestle")
print(response2)
```

```python

response3=query_engine.query("What is Adjusted EBITDA of Uber in 2020 and 2021 ?")
print(response3)
```

```python
def ask_question(query):
  response=query_engine.query(query)
  return response
```

```python
#!pip install gradio
```

```python
#!pip install gradio
```

```python
import gradio as gr
```

```python
# Define the Gradio interface
interface = gr.Interface(
    fn=ask_question,
    inputs=gr.Textbox(lines=2, placeholder="Enter your query here..."),  # Define the input component
    outputs="text",  # Define the output type
    title="Semantic Search Engine",
    description="Enter a query to retrieve information from the documents and get an AI-generated response."
)

# Launch the Gradio interface
interface.launch()
```
